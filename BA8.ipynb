```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```

```python
plt.plot(x,y,"--")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```

```python
plt.plot(x,y,marker="o")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```

```python
plt.plot(x,y,"*")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```

```python
plt.plot(x,y,"r")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```

```python
plt.plot(x,y,"g")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```

```python
plt.plot(x,y,"o:r")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```

```python
plt.plot(x,y,"o:r", ms = 10)
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```

```python
plt.plot(x,y,"o:r", ms = 10, mec = "b")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```
